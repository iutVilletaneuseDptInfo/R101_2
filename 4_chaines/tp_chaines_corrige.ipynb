## Exercice 1 : Est-ce un nombre entier positif ? *

### Question 1

Définir une fonction `estNb` qui prend en paramètre une chaîne de caractères et retourne `True` si elle représente un nombre entier positif. 

**Exemples** :
- `estNb('135')` doit retourner `True`
- `estNb('13A5')` doit retourner `False`
- `estNb('  135')` doit retourner `False`

````python
##################
#   Correction   #
##################


def estNb(s):
    i = 0
    while i < len(s) and s[i] <= '9' and s[i] >= '0':
        i += 1
    return i == len(s)
````

### Question 2

Définir une fonction de tests unitaires pour la fonction `estNb`.

````python
##################
#   Correction   #
##################


def test_estNb():
    assert estNb('135')
    assert not estNb('13A5')
    assert not estNb(' 135')
    assert not estNb('135 ')
    print("Test de la fonction estNb : ok")


test_estNb()
````

## Exercice 2 : Présentation du langage HTML **

Les pages Web sont écrites dans un langage appelé HTML (HyperText Markup Language / langage de marquage hypertexte). C'est un langage de description de documents qui permet de décrire la structure du document et son contenu.  HTML est un langage qui utilise des balises (ou marqueurs, ou tags).  Toute structure de texte est encadrée par une paire de balises.  Le canevas de base d'un document HTML est le suivant :

````python
<html> 
  <head> 
     <title> Exemple pour le cours </title>
  </head>
  <body> 
   <p> C'est ici que l'on écrit le texte de la page </p>
  </body> 
</html>
````

- Chaque balise est reconnaissable car entourée des caractères ``< >``
- Certaines balises vont par *2* : ``<html></html>, <head></head>, <p></p>``. Elles indiquent comment le texte qui est entre la paire de balises doit être structuré. La deuxième balise (la balise de fin) est construite à partir de la première (balise de début) en lui ajoutant devant le caractère ``/``.  
- D'autres balises sont *seules* :``<br/>`` indique un passage à la ligne,``<img/>`` indique l'insertion d'une image. 
- On supposera que le nom des balises HTML, c'est-à-dire la chaîne comprises entre ``<``(ou ``</``) et ``>``, est uniquement composée de *lettres minuscules*. 

L'objectif est de définir différentes fonctions concernant les documents HTML que l'on voit comme une seule chaîne de caractères.

### Question 1

Une balise de début est entourée des caractères ``<`` et ``>`` et entre les deux, elle n'est composée que de **caractères alphabétiques minuscules**.

- Définir une fonction ``estBaliseDebut()`` qui prend en paramètre une chaîne de caractères `s` et vérifie cette chaîne est une balise de début. Elle retournera `True` si c'est le cas  et `False` dans le cas contraire.

````python
##################
#   Correction   #
##################
def estBaliseDebut(s):
    """Verifie si s est une balise de debut"""
    Test=True # vaut True tant que l'on ne rencontre pas d'erreur
    
    if len(s)<2 or s[0] !='<' or s[len(s)-1] != '>':  
        Test=False
    
    # reste à verifier l'interieur
    i = 1
    while i < len(s)-1:
        if 'a'>s[i] or s[i]>'z':
            Test=False
        i+=1
    return Test
````

- Définir une fonction de tests unitaires qui permet de vérifier que la fonction `estBaliseDebut` fait ce qui est attendu.

````python
def test_estBaliseDebut():
    tests = ['<head>', 'head>', '<TITLE>', '<t1t2>', '</head>',""]
    assert estBaliseDebut(tests[0])
    i = 1
    while i < len(tests):
        assert not estBaliseDebut(tests[i])
        i += 1
    print("Test de la fonction estBaliseDebut : ok")


test_estBaliseDebut()
````

### Question 2

- Définir une fonction `construireBaliseFin` qui prend en paramètre une balise de début `balDeb` et retourne un tableau contenant deux valeurs, un booléen et une chaîne de caractères :
    - la première valeur du tableau vaut `True` dans le cas ou la balise `balDeb` est bien une balise de début, auquel cas, la deuxième valeur contenu dans le tableau représente la balise de fin correspondant à `balDeb`,
    - la première valeur du tableau vaut `False` dans le cas ou la balise `balDeb` n'est pas conforme aux balises de début du langage HTML, auquel cas, la deuxième valeur du tableau qui est retourné est une chaîne vide.

````python
##################
#   Correction   #
##################

def construireBaliseFin(s):
    """Construit une balise de fin à partir de s """
    reponse = []
    if not (estBaliseDebut(s)): # s n'est pas une balise de debut
        reponse.append(False)
        reponse.append('')
    else:
        balFin = '</'
        i = 1    # le 1er caractere est `<`, on commence au 2ème
        while i < len(s):   
            balFin += s[i]
            i += 1
        reponse.append(True)
        reponse.append(balFin)
    return reponse
````

- Définir une fonction de tests unitaires qui permet de vérifier que la fonction `construireBaliseFin` fait ce qui est attendu.

````python
def test_construireBaliseFin():
    assert construireBaliseFin('<head>')==[True,'</head>']
    assert construireBaliseFin('<html>')==[True,'</html>']
    assert construireBaliseFin('html>')==[False,'']
    assert construireBaliseFin('<html')==[False,'']
    assert construireBaliseFin('<hTml>')==[False,'']
    print("Test de la fonction construireBaliseFin : ok")

test_construireBaliseFin()
````

## Exercice 3 : Vérification du parenthésage  d'une expression **

### Question 1

Définir une fonction `parentheses()`qui indique si une chaîne de caractères passée en paramètre et représentant une expression parenthésée est syntaxiquement correcte du point de vue des parenthèses ou pas. La fonction renvoie `-1` si l'expression est correcte et la position de la première erreur si l'expression est incorrecte.  

**Exemples**:
- `parentheses("(a.(b))")`  retournera  `-1`.
- `parentheses("(()())")` retournera `-1`.
- `parentheses("a.(b))(")` retournera `5`.
- `parentheses("(()(()a.(b))")` retournera `11`.
- `parentheses(")(")` retournera `0`.

````python
##################
#   Correction   #
##################


def parentheses(s):
    """Verifie que l'expression est bien parenthesee"""
    i = 0
    ouverte = 0

    while i < len(s) and ouverte >= 0:
        if s[i] == '(':
            ouverte = ouverte+1
        elif s[i] == ')':
            ouverte = ouverte-1
        i = i + 1
    if ouverte == 0:
        return -1
    elif ouverte == -1:
        return i-1  # une fermante en trop
    else:
        return i-1  # expression incomplete
````

### Question 2

Définir une fonction de tests unitaires pour la fonction `parentheses`.

````python
##################
#   Correction   #
##################


def test_parentheses():
    expression = '(a.(b))'
    assert parentheses(expression) == -1
    expression = '(()())'
    assert parentheses(expression) == -1
    expression = '(()(()a.(b))'
    assert parentheses(expression) == 11
    expression = ')('
    assert parentheses(expression) == 0
    print("Test de la fonction parentheses : ok")


test_parentheses()
````

    **Exemples**:
    - `parenthesesCrochets("(a.(b))")` retournera  `-1`.
    - `parenthesesCrochets("[a...b]")` retournera `-1`.
    - `parenthesesCrochets("([]())")` retournera `-1`.
    - `parenthesesCrochets("[a(b]c)")` retournera `4`.

    **Remarque :** On pourrait également prendre en compte les accolades.
    
- Définir une fonction de tests unitaires pour la fonction `parenthesesCrochets`.

````python
##################
#   Correction   #
##################


def parentheseCrochets(s):
    pile = []
    fermanteOuvrante = { 
        ")":"(", 
        "]":"["
    }
    ouvrantes = list(fermanteOuvrante.values())
    i = 0
    while i < len(s):
        if s[i] in ouvrantes:
            pile.append(s[i])
        if s[i] in fermanteOuvrante:
            if pile[len(pile)-1] != fermanteOuvrante[s[i]]:
                return i
            else:
                pile.pop()
        i += 1
    if len(pile) > 0:
        return i
    else:
        return -1

def test_parentheseCrochets():
    assert parentheseCrochets("(a.(b))") == -1
    assert parentheseCrochets("[a...b]") == -1
    assert parentheseCrochets("([]())") == -1
    assert parentheseCrochets("[a(b]c)") == 4
    print("Test de la fonction parentheseCrochets : ok")

test_parentheseCrochets()
````

### Question 1

Définir la fonction `paragraph` prenant en paramètre une chaîne de caractères correspondant à un paragraphe en Markdown et retournant ce paragraphe au format HTML.

Par exemple, `paragraph("bonjour")` doit renvoyer `"<p>bonjour</p>"`.

````python
##################
#   Correction   #
##################


def paragraph(s):
    return "<p>" + s + "</p>"
````

### Question 2

Définir la fonction `titleLevel` prenant en paramètre une chaîne de caractères et retournant le niveau du titre, ou 0 si ce n'est pas un titre.

**Exemples :**
- `titleLevel("bonjour")` doit retourner 0
- `titleLevel("#bonjour")` doit retourner 1
- `titleLevel("### bonjour")` doit retourner 3
- `titleLevel("## #bonjo##ur")` doit retourner 2

````python
##################
#   Correction   #
##################


def titleLevel(s):
    i = 0
    while i < len(s) and s[i] == "#":
        i += 1
    return min(i, 6) #niveau max de titre 6
````

### Question 3

Définir la fonction `title` prenant en paramètre une chaîne de caractères et retournant le titre correspondant au format HTML (pas de modification si ce n'est pas un titre).

Par exemple, `title("# Titre de niveau 1")` doit retourner <code>"&lt;h1> Titre de niveau 1&lt;/h1>"</code>.

````python
##################
#   Correction   #
##################


def title(s):
    lev = titleLevel(s)
    if lev == 0:
        return s
    html =  "<h" + str(lev) + ">" 
    i = lev 
    while i < len(s):
        html += s[i]
        i += 1
    return html + "</h" + str(lev) + ">" 
````

### Question 4

Définir la fonction `isItem` prenant en paramètre une chaîne de caractères et retournant `True` si cette chaîne est un item de liste (ie, commence par un tiret), et `False` sinon.

````python
##################
#   Correction   #
##################


def isItem(s):
    return s != "" and s[0] == "-"
````

### Question 5

Définir la fonction `item` prenant en paramètre une chaîne de caractères et retournant l'item sous forme HTML (pas de modification si ce n'est pas un item de liste).

Par exemple, `item("- html")` doit retourner `"<li>html</li>"`.

````python
##################
#   Correction   #
##################


def item(s):
    if isItem(s):
        return "<li>" + s[1:] + "</li>"
    else:
        return s
````

### Question 6

Définir la fonction `htmlLine` prenant un titre, un item de liste ou un paragraphe au format Markdown et retournant la chaîne formatée au format HTML.

````python
##################
#   Correction   #
##################


def htmlLine(s):
    if isItem(s):
        return item(s)
    if titleLevel(s) != 0:
        return title(s)
    return paragraph(s)
````

### Question 7

Définir la fonction `html` prenant en paramètre une chaîne de caractères au format markdown et formattant celle-ci au format html.

**Attention :** Pensez à entourer les items de liste consécutifs dans des balises `ul`.

````python
##################
#   Correction   #
##################


def html(s):
    code = ""
    inUl = False # inUl vaut True si on parse une liste, False sinon
    lignes = s.split("\n")
    i = 0
    while i < len(lignes):
        ligne = lignes[i]
        if ligne != "":
            if isItem(ligne) and not inUl:
                code += "<ul>\n"
                inUl = True
            if not isItem(ligne) and inUl:
                code += "</ul>\n"
                inUl = False
            code += htmlLine(ligne) + "\n"
        i += 1

    if inUl:
        code += "</ul>\n"
    return """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Document</title>
</head>
<body>""" + code + """</body>
</html>"""
````
