## Exercice 2 : Hachage de chaînes selon la somme des codes ASCII de leurs caractères.

On utilise comme table de hachage un tableau `table` où la case d'indice `i` contient l'ensemble des chaînes de caractères dont la somme `somme` des codes ASCII des caractères est égale à `i` modulo la longueur de `table`, autrement dit `somme%len(table)` vaut `i`.

```python
# Fonction de hachage définie en cours

def hachage2(val):
    val = str(val) # pour hacher des valeurs qui ne sont pas des string
    nombre = 0
    i = 0
    while i < len(val):
        nombre += ord(val[i])
        i += 1
    return nombre
```
