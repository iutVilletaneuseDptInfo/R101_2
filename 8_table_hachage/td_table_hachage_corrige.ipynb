## Exercice 1 : Hachage d'un entier selon son dernier chiffre 

On utilise comme table de hachage un tableau de 10 cases où la case d'indice `i` contient l'ensemble des nombres terminant par le chiffre `i`.

#### Exemple de table de hachage

```python
table = [
    [0],
    [3321, 1],
    [12, 59482, 2, 342],
    [23, 543],
    [34, 4, 34534],
    [45, 345],
    [],
    [67, 57, 847, 897],
    [78, 18],
    [7069]
]
```

### Question 1  :  La recherche

- Définir une fonction `rechercher` qui prend en paramètre une table de hachage et un entier et renvoie `True` si l'entier est dans la table, et `False` sinon. 

```python
##################
#   Correction   #
##################

def rechercher(table, entier):
    indice = abs(entier) % 10
    tab = table[indice] # le tableau à la case indice
    
    j = 0
    while j < len(tab):
        if tab[j] == entier:
            return True
        j += 1
    return False
```

### Question 2 : L'ajout  

- Définir une fonction `ajouter` qui prend en paramètre une table de hachage et un entier et ajoute cet entier dans la table s'il n'existe pas. 

```python
##################
#   Correction   #
##################

def ajouter(table, entier):
    if not rechercher(table, entier):
        table[abs(entier)%10].append(entier)
```

### Question  3  : La suppression

- Définir une fonction `supprimer` qui prend en paramètre une table de hachage et un entier et supprime cet entier dans la table s'il est présent.

**Remarque :** On utilisera la fonction `L.pop()` qui supprime le dernier élément de la table `L`.

```python
##################
#   Correction   #
##################

def supprimer(table, entier):
    indice = abs(entier) % 10
    
    # On cherche la position de l'entier dans table[indice]
    j = 0
    while j < len(table[indice]) and table[indice][j] != entier:
        j += 1
        
    # entier est dans la table    
    if j < len(table[indice]):
        #On met le dernier élément de table[indice] à la place de entier
        table[indice][j] = table[indice][len(table[indice]) - 1]
        table[indice].pop()  # On supprime le dernier élément de table[indice]
```

## Exercice 2 : Hachage de chaînes selon la somme des codes ASCII de leurs caractères.

```python
# Fonction de hachage définie en cours

def hachage2(val):
    val = str(val) # pour hacher des valeurs qui ne sont pas des string
    nombre = 0
    i = 0
    while i < len(val):
        nombre += ord(val[i])
        i += 1
    return nombre
```

### Question 3 : La suppression  
- Définir une fonction `supprimer2` qui prend en paramètre une table de hachage et une chaîne de caractères et supprime cette chaîne de caractères dans la table si elle existe. 

**Remarque :** on utilisera la fonction `L.pop()` qui supprime le dernier élément de la table `L`.

```python
##################
#   Correction   #
##################

def supprimer2(table, valeur):
    
    indice = hachage2(valeur) % len(table)
    # On cherche la position de l'entier dans table[indice]
    j = 0
    while j < len(table[indice]) and table[indice][j] != valeur:
        j += 1
    
    if j < len(table[indice]):
        #On met le dernier élément de table[indice] à la place de entier
        table[indice][j] = table[indice][len(table[indice]) - 1]
        table[indice].pop()  # On supprime le dernier élément de table[indice]
```

### Question 4 : Initialisation d'une table de hachage

Créer une fonction `initialiser_table` prenant en paramètre un entier `n` et retournant un tableau contenant `n` cases, chacune d'elle contenant un tableau vide.

```python
##################
#   Correction   #
##################


def initialiser_table(n):
    table = []
    i = 0
    while i < n:
        table.append([])
        i += 1
    return table


t = initialiser_table(5)
print(t)
```
