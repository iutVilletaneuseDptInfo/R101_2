## Partie 2 : Histoires d'anagrammes

#### Tri par ordre alphabétique d'une chaîne de caractères
On suppose que l'on dispose de la fonction `str_sort` qui prend une chaîne de caractères et retourne la chaîne triée (définie à la Question 4 de la Partie 1).

Au besoin, il est possible d'utiliser pour la fonction `str_sort` le code suivant, reposant sur la fonction de tri `sort` et concaténation `join` des éléments d'un tableau de Python : 

```python
def str_sort(s):
    l = list(s)       # Crée un tableau à partir de la chaîne
    l.sort()          # Trie le tableau 
    return "".join(l) # Crée la chaîne de caractères par 
                      # concaténation des caractères du tableau 
```

#### Dictionnaire des anagrammes 
On définit un dictionnaire des anagrammes à partir d'un tableau de mots de la manière suivante  
- les clés sont les chaînes de caractères obtenues en triant chaque mot du tableau dans l'ordre alphabétique ; 
- les valeurs sont des tableaux contenant tous les anagrammes de la clé trouvés dans le tableau de mots (autrement dit, tous les mots du tableau, qui une fois triés par ordre alphabétique, correspondent à la clé).

Par exemple, le tableau de mots est :
```python
["chine", "romain", "manoir", "chien", "chat", "niche", "souris"]
```
le dictionnaire obtenu est le suivant 
```python
{
   "cehin": ["chine", "chien", "niche"], 
   "aimnor": ["romain", "manoir"],
   "acht": ["chat"], 
   "iorssu": ["souris"]
}
```

**Attention :** l'ordre des mots dans les tableaux correspondant aux valeurs du dictionnaire n'a pas d'importance. Dans l'exemple, le dictionnaire pourrait aussi être :
```python
{
   "cehin": ["chien", "chine", "niche"], 
   "aimnor": ["manoir", "romain"],
   "acht": ["chat"], 
   "iorssu": ["souris"]
}
```

- Vérifier que la fonction passe les tests unitaires ci-dessous.

````python
def test_dico_anagrammes():
    mots = ["chine", "romain", "manoir", "chien", "chat", "niche", "souris"]
    expected = {
       "cehin": ["chien", "chine", "niche"], 
       "aimnor": ["manoir", "romain"],
       "acht": ["chat"], 
       "iorssu": ["souris"]
    }
    obtained = dico_anagrammes(mots)
    
    # On trie les tableaux correspondant aux valeurs du dictionnaire pour pouvoir ensuite comparer si ce sont 
    # les mêmes que ceux attendus (déjà triés dans l'ordre)
    cles = list(obtained)
    i = 0
    while i < len(cles):
        obtained[cles[i]].sort()
        i += 1
    assert obtained == expected
    print("Test de la fonction dico_anagrammes : ok")
    
test_dico_anagrammes()    
````
