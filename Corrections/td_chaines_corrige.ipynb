## Exercice 1 : Insertions de symboles * 

- Ecrire une fonction qui prend une chaîne en paramètre et qui renvoie une chaîne dans laquelle ont été insérées des tirets bas (`_`) entre chaque lettre de la chaîne d'origine.
Par exemple, si on lui passe la chaîne `bonjour`, la fonction retourne la chaîne `b_o_n_j_o_u_r`.

- Ecrire une fonction de tests unitaires permettant de vérifier que la fonction fait ce qui est attendu.

```python
##################
#   Correction   #
##################


def tiret(s):
    """Retourne une nouvelle chaîne contenant les caractères de s séparés par des tirets"""
    stiret = ''
    i = 0
    while i < len(s):
        stiret += s[i]
        if i < len(s)-1:
            stiret += '_'
        i = i+1
    return stiret


def test_tiret():
    assert tiret('bonjour') == 'b_o_n_j_o_u_r'
    print("Test de la fonction tiret : ok")


test_tiret()
```

## Exercice 2 : Retirer tous les caractères autres que lettres et espaces *

### Question 1

Définir une fonction `onlySpacesAndLetters` qui prend en paramètre une chaîne de caractères et en retourne une deuxième ne contenant que les lettres (sans accents) et les espaces. 

Par exemple, `onlySpacesAndLetters("Ah, ok, incroyable 2024 !!!")` doit retourner `"Ah ok incroyable  "`.

```python
##################
#   Correction   #
##################


def onlySpacesAndLetters(s):
    ns = ""
    i = 0
    while i < len(s):
        if (s[i] >= "A" and s[i] <= "Z") or (s[i] >= "a" and s[i] <= "z") or s[i] == " ":
            ns += s[i]
        i += 1
    return ns

```

### Question 2

Définir une fonction de tests unitaires pour la fonction `onlySpacesAndLetters`.

```python
##################
#   Correction   #
##################


def test_onlySpacesAndLetters():
    assert onlySpacesAndLetters("Ah, ok, incroyable 2024 !!!") == "Ah ok incroyable  "
    assert onlySpacesAndLetters("1234") == ""
    assert onlySpacesAndLetters("ab c") == "ab c"
    print("Test de la fonction onlySpacesAndLetters : ok")
    
test_onlySpacesAndLetters()
```

## Exercice 3 : Supprimer deux caractères consécutifs identiques **

### Question 1

Définir la fonctio `rmConsecutiveEqualChar` prenant en paramètre une chaîne de caractères et supprimant les caractères identiques consécutifs.

Par exemple, l'appel `rmConsecutiveEqualChar("Ahhh, Ahhh,  AAtchoumm!!")` doit renvoyer `"Ah, Ah, Atchoum!"`. 

```python
##################
#   Correction   #
##################


def rmConsecutiveEqualChar(s):
    if s == "":
        return s
    ns = s[0]
    i = 1
    while i < len(s):
        #Si le ième caractère de s est différent du dernier de ns
        if s[i] != ns[len(ns) - 1]:
            ns += s[i]
        i += 1
    return ns
```

### Question 2

Définir une fonction de tests unitaires pour la fonction `rmConsecutiveEqualChar`.

```python
##################
#   Correction   #
##################


def test_rmConsecutiveEqualChar():
    assert rmConsecutiveEqualChar("") == ""
    assert rmConsecutiveEqualChar("abc") == "abc"
    assert rmConsecutiveEqualChar("aabbcc") == "abc"
    assert rmConsecutiveEqualChar("aaaabc") == "abc"
    assert rmConsecutiveEqualChar("daaaabbbbbcccccd") == "dabcd"
    print("Test de la fonction rmConsecutiveEqualChar : ok")
test_rmConsecutiveEqualChar()
```

## Exercice 4: renversement partiel de chaîne **

### Question 1

Définir la fonction `reverse` prenant en paramètre une chaîne de caractères `s` et retournant une chaîne de caractères contenant les caractères de `s` dans l'ordre inverse.

Par exemple, `reverse("hello")` doit retourner `olleh`.

```python
##################
#   Correction   #
##################


def reverse(s):
    ns = ""
    i = 0
    while i < len(s):
        ns = s[i] + ns
        i += 1
    return ns
```

### Question 2

Définir la fonction `partialReverse` prenant en paramètre une chaîne `s` et deux entiers `i` et `j`, et retournant une chaîne contenant les caractères où seuls ceux compris entre les indices `i` et `j` (inclus) sont renversés.

Par exemple, `partialReverse("hello", 1, 2)` doit retourner `"hlelo"`, alors que `partialReverse("bonjour", 2, 5)` doit retourner `"bouojnr"`.

**Remarque :** `partialReverse(s, 0, len(s) - 1)` doit retourner le même résultat que `reverse(s)`.

```python
##################
#   Correction   #
##################


def partialReverse(s, i, j):
    if i > j:
        i, j = j, i
    i = max(0, i)
    j = min(j, len(s) - 1)
    
    ns = ""
    k = 0
    while k < i:
        ns += s[k]
        k += 1
        
    while k <= j:
        ns += s[i+j-k]
        k += 1
        
    while k < len(s):
        ns += s[k]
        k += 1  
    return ns
```

### Question 3

Définir des fonctions de tests unitaires pour `reverse` et `partialReverse`. 

```python
##################
#   Correction   #
##################


def test_reverse():
    assert reverse("bonjour") == "ruojnob"
    assert reverse("") == ""
    assert reverse("a") == "a"
    print("Test de la fonction reverse : ok")

def test_partialReverse():
    assert partialReverse("hello", 1, 2) == "hlelo"
    assert partialReverse("bonjour", 2, 5) == "bouojnr"
    assert partialReverse("bonjour", 5, 5) == "bonjour"
    assert partialReverse("bonjour", 5, 6) == "bonjoru"
    assert partialReverse("bonjour", 0, 6) == reverse("bonjour")
    print("Test de la fonction partialReverse : ok")

test_reverse()
test_partialReverse()
```
