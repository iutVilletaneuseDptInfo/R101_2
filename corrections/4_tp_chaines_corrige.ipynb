## Exercice 1 : Est-ce un nombre entier positif ? *

### Question 1

Définir une fonction `estNb` qui prend en paramètre une chaîne de caractères et retourne `True` si elle représente un nombre entier positif. 

**Exemples** :
- `estNb('135')` doit retourner `True`
- `estNb('13A5')` doit retourner `False`
- `estNb('  135')` doit retourner `False`

````python
##################
#   Correction   #
##################


def estNb(s):
    i = 0
    while i < len(s) and s[i] <= '9' and s[i] >= '0':
        i += 1
    return i == len(s)
````

### Question 2

Définir une fonction de tests unitaires pour la fonction `estNb`.

````python
##################
#   Correction   #
##################


def test_estNb():
    assert estNb('135')
    assert not estNb('13A5')
    assert not estNb(' 135')
    assert not estNb('135 ')
    print("Test de la fonction estNb : ok")


test_estNb()
````

## Exercice 2 : Présentation du langage HTML **

Les pages Web sont écrites dans un langage appelé HTML (HyperText Markup Language / langage de marquage hypertexte). C'est un langage de description de documents qui permet de décrire la structure du document et son contenu.  HTML est un langage qui utilise des balises (ou marqueurs, ou tags).  Toute structure de texte est encadrée par une paire de balises.  Le canevas de base d'un document HTML est le suivant :

````python
<html> 
  <head> 
     <title> Exemple pour le cours </title>
  </head>
  <body> 
   <p> C'est ici que l'on écrit le texte de la page </p>
  </body> 
</html>
````

Chaque balise est reconnaissable car entourée des caractères ``< >``

Certaines balises vont par 2 : ``<html></html>, <head></head>, <p></p>``. Elles indiquent comment le texte qui est entre la paire de balises doit être structuré. La deuxième balise (la balise de fin) est construite à partir de la première (balise de début) en lui ajoutant devant le caractère ``/``.  D'autres balises sont *seules* :``<br/>`` indique un passage à la ligne,``<img/>`` indique l'insertion d'une image. On supposera que le nom des balises HTML, c'est-à-dire la chaîne comprises entre ``<``(ou ``</``) et ``>``, est uniquement composée de lettres minuscules. L'objectif est de éfinir différentes fonctions concernant les documents HTML.  On suppose que le document est représenté par une chaîne de caractères :

### Question 1

Définir une fonction ``estBaliseDebut()`` qui vérifie si une chaîne ``s``, reçue en argument est une balise de début.  Une balise de début est entourée des caractères ``<`` et ``>`` et entre les deux, elle n'est composée que de **caractères alphabétiques minuscules**.


````python
##################
#   Correction   #
##################


def isalpha(c):
    """Verifie si un caractere alphabetique est minuscule"""
    return 'a' <= c and c <= 'z'


def estBaliseDebut(s):
    """Verifie si s est une balise de debut"""
    lg = len(s)
    if s[0] == '<' and s[lg-1] == '>':  # reste à verifier l'interieur
        i = 1
        while i < lg-1 and isalpha(s[i]):
            i = i + 1
        return i == lg-1
    else:
        return False


def test_estBaliseDebut():
    tests = ['<head>', 'head>', '<TITLE>', '<t1t2>', '</head>']
    assert estBaliseDebut(tests[0])
    i = 1
    while i < len(tests):
        assert not estBaliseDebut(tests[i])
        i += 1
    print("Test de la fonction estBaliseDebut : ok")


test_estBaliseDebut()
````

### Question 2

Définir une fonction ``construireBaliseFin(balDeb)`` qui retourne un tableau contenant deux valeurs, un booléen et une chaîne de caractères :
- la première valeur du tableau vaut `True` dans le cas ou la balise `balDeb` est une balise de début, auquel cas, la deuxième valeur contenu dans le tableau représente la balise de fin correspondant à `balDeb`,
- la première valeur du tableau vaut `False` dans le cas ou la balise `balDeb` n'est pas conforme au langage HTML, auquel cas, la deuxième valeur du tableau qui est retourné est une chaîne vide.


````python
##################
#   Correction   #
##################


def construireBaliseFin(s):
    """Construit une balise de fin a partir de s """
    reponse = []
    if not (estBaliseDebut(s)):  # s n'est pas une balise de debut
        reponse.append(False)
        reponse.append('')
        return reponse  # on retourne False et une chaine vide
    else:
        balFin = '</'
        i = 1
        while i < len(s):  # le premier caractere est l'ouverture de la balise '<'
            balFin = balFin+s[i]
            i = i+1
        reponse.append(True)
        reponse.append(balFin)
        return reponse


def test_construireBaliseFin():
    bal = '<head>'
    rep = construireBaliseFin(bal)
    assert rep[0] == True and rep[1] == '</head>'
    bal = 'toto>'
    rep = construireBaliseFin(bal)
    assert rep[0] == False and rep[1] == ''
    print("Test de la fonction construireBaliseFin : ok")


test_construireBaliseFin()
````

## Exercice 3 : Vérification du parenthésage  d'une expression **

### Question 1

Définir une fonction `parentheses()`qui indique si une chaîne de caractères passée en praramètre et représentant une expression parenthésée est syntaxiquement correcte du point de vue des parenthèses ou pas. La fonction renvoie `-1` si l'expression est correcte et la position de la première erreur si l'expression est incorrecte.  

**Exemples**:
- pour "(a.(b))" elle retourne  -1
- pour "(()())" elle retourne -1
- pour "a.(b))(" elle retourne 5
- pour "(()(()a.(b))" elle retourne 12
- pour ")(" elle retourne 0

````python
##################
#   Correction   #
##################


def parentheses(s):
    """Verifie que l'expression est bien parenthesee"""
    i = 0
    ouverte = 0

    while i < len(s) and ouverte >= 0:
        if s[i] == '(':
            ouverte = ouverte+1
        elif s[i] == ')':
            ouverte = ouverte-1
        i = i + 1
    if ouverte == 0:
        return -1
    elif ouverte == -1:
        return i-1  # une fermante en trop
    else:
        return i  # expression incomplete
````

### Question 2

Définir une fonction de tests unitaires pour la fonction `parentheses`.

````python
##################
#   Correction   #
##################


def test_parentheses():
    expression = '(a.(b))'
    assert parentheses(expression) == -1
    expression = '(()())'
    assert parentheses(expression) == -1
    expression = '(()(()a.(b))'
    assert parentheses(expression) == 12
    expression = ')('
    assert parentheses(expression) == 0
    print("Test de la fonction parentheses : ok")


test_parentheses()
````

La fonction renvoie `-1` si l'expression est correcte et la position de la première erreur si l'expression est incorrecte.  

**Exemples**:
- pour "(a.(b))" elle retourne  -1
- pour "[a...b]" elle retourne -1
- pour "([]())" elle retourne -1
- pour "[a(b]c)" elle retourne 4

**Remarque :** vous pouvez également prendre en compte les accolades.


````python
##################
#   Correction   #
##################


def parentheseCrochets(s):
    pile = []
    fermanteOuvrante = { 
        ")":"(", 
        "]":"["
    }
    ouvrantes = list(fermanteOuvrante.values())
    i = 0
    while i < len(s):
        if s[i] in ouvrantes:
            pile.append(s[i])
        if s[i] in fermanteOuvrante:
            if pile[len(pile)-1] != fermanteOuvrante[s[i]]:
                return i
            else:
                pile.pop()
        i += 1
    if len(pile) > 0:
        return i
    else:
        return -1

def test_parentheseCrochets():
    assert parentheseCrochets("(a.(b))") == -1
    assert parentheseCrochets("[a...b]") == -1
    assert parentheseCrochets("([]())") == -1
    assert parentheseCrochets("[a(b]c)") == 4
    print("Test de la fonction parentheseCrochets : ok")

test_parentheseCrochets()
````

## Exercice 4: From markdown to html ***

Le langage [https://fr.wikipedia.org/wiki/Markdown](Markdown) est un langage de balisage léger qui permet d'écrire un texte avec quelques formattages.

On souhaite créer une fonction qui prend un texte au format mardown et le transforme en html. On considèrera uniquement les formattages markdown suivants :

### Question 1

Définir la fonction `paragraph` prenant en paramètre une chaîne de caractères correspondant à un paragraphe en markdown et retournant ce paragraphe au format html.

Par exemple, `paragraph("bonjour")` doit renvoyer `"<p>bonjour</p>"`.

````python
##################
#   Correction   #
##################


def paragraph(s):
    return "<p>" + s + "</p>"
````

### Question 2

Définir la fonction `titleLevel` prenant en paramètre une chaîne de caractères et retournant le niveau du titre, ou 0 si ce n'est pas un titre.

**Exemples :**
- `titleLevel("bonjour")` doit retourner 0
- `titleLevel("#bonjour")` doit retourner 1
- `titleLevel("### bonjour")` doit retourner 3
- `titleLevel("## #bonjo##ur")` doit retourner 2

````python
##################
#   Correction   #
##################


def titleLevel(s):
    i = 0
    while i < len(s) and s[i] == "#":
        i += 1
    return min(i, 6) #niveau max de titre 6
````

### Question 3

Définir la fonction `title` prenant en paramètre une chaîne de caractères et retournant le titre correspondant au format html (pas de modification si ce n'est pas un titre).

Par exemple `title("# Titre de niveau 1")` doit retourner <code>"&lt;h1> Titre de niveau 1</h1>"</code>.

````python
##################
#   Correction   #
##################


def title(s):
    lev = titleLevel(s)
    if lev == 0:
        return s
    html =  "<h" + str(lev) + ">" 
    i = lev 
    while i < len(s):
        html += s[i]
        i += 1
    return html + "</h" + str(lev) + ">" 
````

### Question 4

Définir la fonction `isItem` prenant en paramètre une chaîne de caractères et retournant `True` si cette chaîne est un item de liste (ie, commence par un tiret), et `False` sinon.

````python
##################
#   Correction   #
##################


def isItem(s):
    return s != "" and s[0] == "-"
````

### Question 5

Définir la fonction `item` prenant en paramètre une chaîne de caractères et retournant l'item sous forme html (pas de modification si ce n'est pas un item de liste).

Par exemple, `item("- html")` doit retourner `"<li>html</li>"`.

````python
##################
#   Correction   #
##################


def item(s):
    if isItem(s):
        return "<li>" + s[1:] + "</li>"
    else:
        return s
````

### Question 6

Définir la fonction `htmlLine` prenant un titre, un item de liste ou un paragraphe au format markdown et retournant la chaîne formattée au format html.

````python
##################
#   Correction   #
##################


def htmlLine(s):
    if isItem(s):
        return item(s)
    if titleLevel(s) != 0:
        return title(s)
    return paragraph(s)
````

### Question 6

Définir la fonction `html` prenant en paramètre une chaîne de caractères au format markdown et formattant celle-ci au format html.

**Attention :** Pensez à entourer les items de liste consécutifs dans des balises `ul`.

````python
##################
#   Correction   #
##################


def html(s):
    code = ""
    inUl = False # inUl vaut True si on parse une liste, False sinon
    lignes = s.split("\n")
    i = 0
    while i < len(lignes):
        ligne = lignes[i]
        if ligne != "":
            if isItem(ligne) and not inUl:
                code += "<ul>\n"
                inUl = True
            if not isItem(ligne) and inUl:
                code += "</ul>\n"
                inUl = False
            code += htmlLine(ligne) + "\n"
        i += 1

    if inUl:
        code += "</ul>\n"
    return """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Document</title>
</head>
<body>""" + code + """</body>
</html>"""
````
